```python
%matplotlib notebook
import matplotlib.pyplot as plt
import pandas as pd
```

```python
i7 = pd.read_csv("i7.csv")
i5 = pd.read_csv("i5.csv")
```

```python
i7.head()
```

```python
i5.head()
```

```python
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(9,6),tight_layout=True)
ax.set_title('Tiempo Medio de Procesamiento de 64 Muestras')
ax.set_ylabel('Tiempo de Procesamiento (s)')
ax.set_xlabel('Número de Pistas de Audio Mono 96KHz 32 bit')
plt.plot(i5["Tracks"], i5["ProcessingTimeAvgSecs"], c='blue',alpha=0.5,label='i5')
plt.plot(i7["Tracks"], i7["ProcessingTimeAvgSecs"], c='red',alpha=0.5,label='i7')
plt.plot(i7["Tracks"], i7["ProcessingTimeLimitSecs"], c='black',alpha=0.5,label='Límite')
plt.xticks(rotation=90,fontsize=6)
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(9,6),tight_layout=True)
ax.set_title('Uso de CPU')
ax.set_ylabel('Uso de CPU (%)')
ax.set_xlabel('Número de Pistas de Audio Mono 96KHz 32 bit')
plt.plot(i5["Tracks"], i5["CPUPer"]*100.0, c='blue',alpha=0.5,label='i5')
plt.plot(i7["Tracks"], i7["CPUPer"]*100.0, c='red',alpha=0.5,label='i7')
plt.xticks(rotation=90,fontsize=6)
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(ncols=1, nrows=1, figsize=(9,6),tight_layout=True)
ax.set_title('Uso de Memoria Residente')
ax.set_ylabel('Memoria Residente (MB)')
ax.set_xlabel('Número de Pistas de Audio Mono 96KHz 32 bit')
plt.plot(i5["Tracks"], i5["ResidentMemoryMB"], c='blue',alpha=0.5,label='i5')
plt.plot(i7["Tracks"], i7["ResidentMemoryMB"], c='red',alpha=0.5,label='i7')
plt.xticks(rotation=90,fontsize=6)
ax.legend()
plt.show()
```
